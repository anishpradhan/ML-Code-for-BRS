```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
```

```python
books = pd.read_csv('../Book_reviews/BX-Books.csv', sep=';', error_bad_lines=False, encoding="latin-1")
books.columns = ['ISBN', 'bookTitle', 'bookAuthor', 'yearOfPublication', 'publisher', 'imageUrlS', 'imageUrlM', 'imageUrlL']
users = pd.read_csv('../Book_reviews/BX-Users.csv', sep=';', error_bad_lines=False, encoding="latin-1")
users.columns = ['userID', 'Location', 'Age']
ratings = pd.read_csv('../Book_reviews/BX-Book-Ratings.csv', sep=';', error_bad_lines=False, encoding="latin-1")
ratings.columns = ['userID', 'ISBN', 'bookRating']
```

```python
top10 = pd.DataFrame(ratings.groupby(['ISBN'])['bookRating'].count())
top10rated = top10.sort_values('bookRating', ascending=False).head(10)
```

```python
new = top10rated.merge(books,left_index = True, right_on='ISBN')
```

```python
plt.rc("font", size = 20)
ratings.bookRating.value_counts(sort=False).plot(kind='bar')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
# average_rating['ratingCount'] = pd.DataFrame(ratings.groupby)
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_rating=average_rating.sort_values('ratingCount', ascending=False).head(10)
```

```python
average_rating.merge(books, left_index=True, right_on='ISBN')
```

```python
counts1 = ratings['userID'].value_counts()
ratings = ratings[ratings['userID'].isin(counts1[counts1 >=200].index)]
counts = ratings['bookRating'].value_counts()
ratings = ratings[ratings['bookRating'].isin(counts[counts >=100].index)]
ratings.head()
```

```python
ratings_pivot = ratings.pivot(index = 'userID', columns='ISBN').bookRating
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
```

```python
print(ratings_pivot.shape)
ratings_pivot
```

```python
def recommendBooks(item_id):
    bones_ratings = ratings_pivot[str(item_id)]
    similar_to_bones = ratings_pivot.corrwith(bones_ratings)
    corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
    corr_bones.dropna(inplace=True)
    corr_summary = corr_bones.join(average_rating['ratingCount'])
    corr_summary = corr_summary[corr_summary['ratingCount']>=300].sort_values('pearsonR', ascending=False).head(10)
    books_corr_to_bones = corr_summary.merge(books, left_index=True, right_on = 'ISBN')
    return books_corr_to_bones['bookTitle']
```
